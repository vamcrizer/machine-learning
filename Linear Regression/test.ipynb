```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Hm s hai bin
def f(x, y):
    return x**2 + 10*y**2

# o hm ring cp hai
def dxx(x, y):
    return 2

def dxy(x, y):
    return 0

def dyy(x, y):
    return 20

# To li ta 
x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)
X, Y = np.meshgrid(x, y)

# Tnh gi tr hm s ti cc im li
Z = f(X, Y)

# To ma trn Hessian
H = np.zeros((30, 30, 2, 2))
H[:, :, 0, 0] = dxx(X, Y)
H[:, :, 0, 1] = dxy(X, Y)
H[:, :, 1, 0] = dxy(X, Y)
H[:, :, 1, 1] = dyy(X, Y)

# Tnh  cong
curvature = np.linalg.det(H) / (1 + np.linalg.norm(np.gradient(Z), axis=0)**2)**(3/2)

# V  th 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
ax.set_title('Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
